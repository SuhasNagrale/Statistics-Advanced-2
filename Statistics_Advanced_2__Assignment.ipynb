# 1. What is hypothesis testing in statistics?

---

**Hypothesis testing** is a **statistical method** used to make decisions or draw conclusions about a **population** based on **sample data**.

It helps you determine whether there is enough evidence in your sample to **support or reject a claim** (called a *hypothesis*) about the population.

---

### **Key Concepts**

| Term                                  | Meaning                                                                                                         |
| ------------------------------------- | --------------------------------------------------------------------------------------------------------------- |
| **Null Hypothesis (H₀)**              | The default assumption — usually states that **no effect**, **no difference**, or **no relationship** exists.   |
| **Alternative Hypothesis (H₁ or Ha)** | The claim you want to test — usually states that **an effect**, **difference**, or **relationship** does exist. |
| **Test Statistic**                    | A value calculated from sample data that measures how far the data is from what we’d expect under H₀.           |
| **p-value**                           | The probability of obtaining results as extreme as (or more extreme than) those observed, if H₀ is true.        |
| **Significance Level (α)**            | The threshold for deciding whether to reject H₀ — commonly set at 0.05 (5%).                                    |

---

###  **Steps in Hypothesis Testing**

1. **State hypotheses:**

   * H₀: The population mean = 50
   * H₁: The population mean ≠ 50

2. **Choose significance level (α):**
   e.g., α = 0.05

3. **Select the appropriate test:**
   (e.g., z-test, t-test, chi-square test, ANOVA, etc.)

4. **Compute the test statistic** using sample data.

5. **Find the p-value** and compare it with α.

   * If **p ≤ α**, reject H₀.
   * If **p > α**, fail to reject H₀.

6. **Draw a conclusion** in context of the problem.

---

### **Example**

A data scientist wants to test if the average time users spend on a website is **different from 10 minutes**.

* H₀: μ = 10
* H₁: μ ≠ 10
* α = 0.05
* Sample mean = 11, sample size = 30, s = 2.

After running a **t-test**, you get p = 0.03.
Since **0.03 < 0.05**, you **reject H₀** and conclude that **average time differs from 10 minutes**.

---

### **In Data Science Context**

Hypothesis testing is widely used for:

* Comparing model performances (e.g., A/B testing)
* Evaluating experimental results
* Validating assumptions about datasets
* Feature significance in regression models

# 2. What is the null hypothesis, and how does it differ from the alternative hypothesis?

---

### **1. Null Hypothesis (H₀)**

The **null hypothesis** is a **default or baseline assumption** that there is **no effect, no difference, or no relationship** in the population.
It represents the idea of **“nothing new is happening.”**

 In other words, it assumes that **any observed difference in the sample is due to random chance**.

**Examples:**

* H₀: The average customer rating before and after a website redesign is **the same**.
* H₀: There is **no correlation** between hours studied and exam score.
* H₀: The mean weight of apples = 150 grams.

---

###  **2. Alternative Hypothesis (H₁ or Ha)**

The **alternative hypothesis** is what you **want to test or prove**.
It claims that there **is** an effect, difference, or relationship — **something has changed**.

**Examples:**

* H₁: The average customer rating **changed** after the redesign.
* H₁: There **is** a correlation between hours studied and exam score.
* H₁: The mean weight of apples **is not** 150 grams.

---

### **3. The Key Difference**

| Feature  | **Null Hypothesis (H₀)**                    | **Alternative Hypothesis (H₁ / Ha)** |
| -------- | ------------------------------------------- | ------------------------------------ |
| Meaning  | No effect or difference                     | There is an effect or difference     |
| Role     | The default assumption                      | The research claim                   |
| Decision | We try to **reject** it (not prove it true) | Supported if H₀ is rejected          |
| Symbol   | H₀                                          | H₁ or Ha                             |
| Example  | μ = 50                                      | μ ≠ 50                               |

---

###  **4. Decision Logic**

After performing a statistical test:

* If **p-value ≤ α** (e.g., 0.05) → **Reject H₀**, accept H₁.
* If **p-value > α** → **Fail to reject H₀** (not enough evidence to support H₁).

---

### **5. Real-World Example (A/B Testing)**

A data scientist wants to check if a new website layout increases click-through rate (CTR).

* H₀: CTR_new = CTR_old → No difference in performance.
* H₁: CTR_new > CTR_old → New layout performs better.

If testing shows a statistically significant increase, H₀ is rejected in favor of H₁.

# 3. Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.

---

###  **What Is the Significance Level (α)?**

The **significance level**, denoted by **α (alpha)**, is the **threshold probability** used to decide whether the observed results are **statistically significant** — i.e., unlikely to have occurred by random chance.

It represents the **maximum risk** you are willing to take of **rejecting a true null hypothesis** (also called a **Type I error**).

---

###  **In Simple Terms**

It answers the question:

> “How sure do we want to be before we reject the null hypothesis?”

For example:

* If **α = 0.05**, you are willing to accept a **5% chance** of being wrong when you reject H₀.

---

###  **Common Values of α**

| Significance Level (α) | Confidence Level | Interpretation                                     |
| ---------------------- | ---------------- | -------------------------------------------------- |
| 0.10                   | 90%              | Tolerate a higher risk of Type I error             |
| 0.05                   | 95%              | Most common — balance between risk and reliability |
| 0.01                   | 99%              | Very strict — only reject H₀ with strong evidence  |

---

###  **Role in Hypothesis Testing**

After you compute the **p-value** from your statistical test:

* If **p ≤ α** → **Reject H₀**
  (The result is *statistically significant* — strong evidence against H₀.)

* If **p > α** → **Fail to reject H₀**
  (The result is *not statistically significant* — insufficient evidence to reject H₀.)

---

###  **Example**

Suppose you’re testing whether a new drug lowers blood pressure compared to the standard drug.

* H₀: The new drug has the same effect as the standard drug.
* α = 0.05
* p-value = 0.03

Since **p = 0.03 < 0.05**, you **reject H₀** and conclude that the new drug has a **significant effect**.

---

### **Connection to Errors**

| Type of Error         | Description                  | Controlled by                   |
| --------------------- | ---------------------------- | ------------------------------- |
| **Type I Error (α)**  | Rejecting a true H₀          | Significance level              |
| **Type II Error (β)** | Failing to reject a false H₀ | Sample size, effect size, and α |

So, setting α lower (e.g., 0.01) reduces the chance of a **Type I error**, but may increase the chance of a **Type II error**.

---

### **Summary**

* **α (significance level)** = probability of rejecting H₀ when it’s actually true.
* Determines the **cutoff for statistical significance**.
* Helps decide whether to **reject or fail to reject** the null hypothesis.
* Typical choices: **0.05**, **0.01**, or **0.10**.

---



# 4. What are Type I and Type II errors? Give examples of each.



---

##  **Type I and Type II Errors in Hypothesis Testing**

When we perform a hypothesis test, we decide whether to **reject** or **not reject** the **null hypothesis (H₀)**.
Sometimes, our decision can be wrong — those mistakes are called **Type I** and **Type II errors**.

---

### 🔹 **Type I Error (False Positive)**

* **Definition:**
  Occurs when we **reject H₀** even though it is actually **true**.
  → We think there’s an effect or difference, but in reality, there isn’t.

* **Symbol:** α (alpha)
  It represents the **significance level** of the test (e.g., α = 0.05 means a 5% chance of committing this error).

* **Example:**
  A medical test says a person **has a disease** when they actually **do not**.

  * H₀: The person is healthy.
  * H₁: The person is sick.
    Rejecting H₀ when the person is actually healthy → **Type I error**.

---

### 🔹 **Type II Error (False Negative)**

* **Definition:**
  Occurs when we **fail to reject H₀** even though it is actually **false**.
  → We miss detecting a real effect.

* **Symbol:** β (beta)

* **Example:**
  A medical test says a person **does not have a disease** when they actually **do**.

  * H₀: The person is healthy.
  * H₁: The person is sick.
    Failing to reject H₀ when the person is actually sick → **Type II error**.

---

###  **Summary Table**

| Error Type      | Description                  | Common Name    | Example                             |
| --------------- | ---------------------------- | -------------- | ----------------------------------- |
| **Type I (α)**  | Rejecting a true H₀          | False Positive | Test shows a healthy person is sick |
| **Type II (β)** | Failing to reject a false H₀ | False Negative | Test shows a sick person is healthy |

---

###  **Simple Analogy**

Think of a **court trial**:

* H₀: The defendant is innocent.
* H₁: The defendant is guilty.
  | Decision | True Situation | Result |
  |-----------|----------------|--------|
  | Convict innocent person | Innocent | **Type I error** |
  | Free guilty person | Guilty | **Type II error** |

# 5. What is the difference between a Z-test and a T-test? Explain when to use each.

---

##  **Z-test vs. T-test: Overview**

Both **Z-tests** and **T-tests** are **parametric hypothesis tests** used to compare means —
but they differ mainly in **sample size** and **knowledge of population standard deviation (σ)**.

---

###  **1. Z-Test**

####  **Definition:**

A **Z-test** is used to determine whether there is a **significant difference** between sample and population means (or between two samples) **when the population standard deviation (σ) is known**.

####  **Key Features:**

| Aspect                            | Z-Test                                          |
| --------------------------------- | ----------------------------------------------- |
| Population standard deviation (σ) | Known                                           |
| Sample size (n)                   | Large (usually **n > 30**)                      |
| Distribution                      | **Normal (Z) distribution**                     |
| Test statistic                    | ( Z = \frac{\bar{X} - \mu}{\sigma / \sqrt{n}} ) |

####  **Example:**

A factory claims that the average weight of its packaged snacks is **50 g**.
You take a sample of 100 packets, with a mean weight of 49.5 g and a **known population σ = 2 g**.
You would use a **Z-test** to check if the mean weight differs from 50 g.

---

###  **2. T-Test**

####  **Definition:**

A **T-test** is used to determine whether there is a **significant difference** between sample and population means (or between two samples) **when the population standard deviation (σ) is unknown**.

####  **Key Features:**

| Aspect                            | T-Test                                                              |
| --------------------------------- | ------------------------------------------------------------------- |
| Population standard deviation (σ) | **Unknown**                                                         |
| Sample size (n)                   | **Small (n ≤ 30)**                                                  |
| Distribution                      | **Student’s t-distribution** (broader tails than normal)            |
| Test statistic                    | ( t = \frac{\bar{X} - \mu}{s / \sqrt{n}} ), where **s** = sample SD |

####  **Example:**

You test a new teaching method on **20 students**, and you don’t know the population standard deviation.
You would use a **T-test** to check if their mean score differs from the population average.

---

###  **3. Summary Table**

| Feature           | **Z-Test**                                      | **T-Test**                                 |
| ----------------- | ----------------------------------------------- | ------------------------------------------ |
| Population SD (σ) | Known                                           | Unknown                                    |
| Sample Size       | Large (n > 30)                                  | Small (n ≤ 30)                             |
| Distribution Used | Normal (Z)                                      | Student’s t                                |
| Test Statistic    | ( Z = \frac{\bar{X} - \mu}{\sigma / \sqrt{n}} ) | ( t = \frac{\bar{X} - \mu}{s / \sqrt{n}} ) |
| Example           | Population σ = 10 known                         | σ unknown, only sample data available      |

---

###  **When to Use**

| Situation                                               | Test to Use            |
| ------------------------------------------------------- | ---------------------- |
| Population standard deviation is known                  | **Z-test**             |
| Population standard deviation is unknown and n is small | **T-test**             |
| Comparing two sample means (independent groups)         | **Independent t-test** |
| Comparing the same group before/after treatment         | **Paired t-test**      |

---

###  **In Practice (Data Science Context)**

* **Z-test** → often used in **A/B testing** or **large datasets** where σ is known or can be approximated.
* **T-test** → used when working with **small samples**, e.g., pilot studies, experimental results, or model performance differences on small datasets.

# 7. Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

Let’s assume:

Null hypothesis (H₀): μ = 50

Alternative hypothesis (H₁): μ ≠ 50

Significance level (α): 0.05

Population standard deviation (σ) is known = 0.5 (a common assumption for Z-tests).

```python
import numpy as np
from scipy import stats

# Given data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Known population parameters
mu_0 = 50      # hypothesized population mean
sigma = 0.5    # known population standard deviation

# Calculate sample statistics
sample_mean = np.mean(sample_data)
n = len(sample_data)

# Compute Z-statistic
z_stat = (sample_mean - mu_0) / (sigma / np.sqrt(n))

# Compute two-tailed p-value
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

# Display results
print(f"Sample Mean = {sample_mean:.3f}")
print(f"Z-statistic = {z_stat:.3f}")
print(f"P-value = {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print(" Reject the null hypothesis: The sample mean is significantly different from 50.")
else:
    print(" Fail to reject the null hypothesis: No significant difference from 50.")
```

# Interpretation

The p-value (≈ 0.29) is greater than 0.05, so we fail to reject H₀.

This means there is no statistically significant evidence that the sample mean differs from 50.

In other words, the sample data is consistent with the population mean being 50.

# 8. Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# --------------------------
# Step 1: Simulate normal data
# --------------------------
np.random.seed(42)      # For reproducibility
mu = 50                 # Population mean
sigma = 5               # Population standard deviation
n = 100                 # Sample size

data = np.random.normal(mu, sigma, n)

# --------------------------
# Step 2: Calculate sample statistics
# --------------------------
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)           # Sample standard deviation
std_error = sample_std / np.sqrt(n)

# --------------------------
# Step 3: Calculate 95% Confidence Interval
# --------------------------
confidence = 0.95
t_critical = stats.t.ppf((1 + confidence) / 2, df=n-1)  # For t-distribution
margin_of_error = t_critical * std_error
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

print(f"Sample Mean = {sample_mean:.2f}")
print(f"95% Confidence Interval = ({ci_lower:.2f}, {ci_upper:.2f})")

# --------------------------
# Step 4: Plot histogram with CI
# --------------------------
plt.figure(figsize=(8,5))
plt.hist(data, bins=15, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(sample_mean, color='red', linestyle='--', label=f"Mean = {sample_mean:.2f}")
plt.axvline(ci_lower, color='green', linestyle='--', label=f"95% CI Lower = {ci_lower:.2f}")
plt.axvline(ci_upper, color='green', linestyle='--', label=f"95% CI Upper = {ci_upper:.2f}")
plt.title('Normal Distribution with 95% Confidence Interval')
plt.xlabel('Data Values')
plt.ylabel('Frequency')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

# Explanation

Simulate Data

np.random.normal(mu, sigma, n) generates n random samples from a normal distribution with mean mu and standard deviation sigma.

Sample Statistics

sample_mean → mean of the simulated data.

sample_std → sample standard deviation (ddof=1 ensures unbiased estimate).

std_error = sample_std / sqrt(n) → standard error of the mean.

Confidence Interval

Uses t-distribution (stats.t.ppf) since population standard deviation is unknown.

margin_of_error = t_critical * std_error gives the range around the sample mean.

Plotting

Histogram of the simulated data.

Red dashed line → sample mean.

Green dashed lines → lower and upper bounds of 95% CI.

This gives both a numerical and visual understanding of the mean and its confidence interval.

# 9. Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.

```python
import numpy as np
import matplotlib.pyplot as plt

# --------------------------
# Function to calculate Z-scores
# --------------------------
def calculate_z_scores(data):
    """
    Calculate Z-scores for a dataset.

    Z-score formula: Z = (x - mean) / std
    """
    mean = np.mean(data)
    std = np.std(data, ddof=1)  # sample standard deviation
    z_scores = (data - mean) / std
    return z_scores

# --------------------------
# Sample dataset
# --------------------------
data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
        50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5]

# --------------------------
# Calculate Z-scores
# --------------------------
z_scores = calculate_z_scores(data)
print("Z-scores:\n", z_scores)

# --------------------------
# Plot histogram of Z-scores
# --------------------------
plt.figure(figsize=(8,5))
plt.hist(z_scores, bins=10, color='lightblue', edgecolor='black', alpha=0.7)
plt.axvline(0, color='red', linestyle='--', label='Mean (Z=0)')
plt.title('Histogram of Z-scores')
plt.xlabel('Z-score')
plt.ylabel('Frequency')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

# Explanation of Z-scores

A Z-score indicates how many standard deviations a data point is away from the mean.

Formula:
Z = (X-mean)/std
	​


Interpretation:

Z = 0 → data point equals the mean.

Z = 1 → 1 standard deviation above the mean.

Z = -1 → 1 standard deviation below the mean.

Helps standardize data, compare across different datasets, and detect outliers.

In the histogram:

Most Z-scores will be between -2 and 2 for a roughly normal distribution.

Red dashed line at Z=0 shows the mean.
